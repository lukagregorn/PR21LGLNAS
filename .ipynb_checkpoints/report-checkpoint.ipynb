### PREDSTAVITEV MNOICE PODATKOV 
-----

Naa mnoica podatkov, nad katero bomo izvajali raziskave je podatkovna baza sporoil iz Discord strenika, ki smo ga tudentje ustvarili za medsebojno pomo. Trenutno je na serverju ve kot 1500 tudentov vkljuenih v enega od programov FRI, ki dnevno sodeluje v pogovorih na razlinih temah. Prvo sporoilo sega vse do leta 2017. 

Za to mnoico podatkov smo se odloili, saj se nam je zdelo, da bi bila ta tema zanimiva za nas ter ostale tudente, ki sodelujejo pri predmetu Podatkovno rudarjenje. 

Podatkovna mnoica je sestavljena iz ve .JSON datotek (za vsak predmet svoja). Vsaka datoteka vsebuje sporoila, ki imajo sledee atribute:

Atribut | Opis      | Tip  |
--------|-----------|-------
id | zaporedna tevilka sporoila | Int
type | tip sporoila | String
timestamp | asovna znaka, kdaj je bilo sporoilo poslano | Time stamp 
timestampEdited | asovna znaka, kdaj je bilo sporoilo popravljeno | Time stamp 
callEndedTimestamp | asovna znaka, kdaj je bil konan klic  | Time stamp 
isPinned | indikator ali je sporoilo pripeto | Boolean
content | vsebina sporoila | String
author | avtor sporoila (id, ime) | Dictionary
attachments | priponke sporoilu (id, url) | Dictionary
embeds | vpeta sporoila (id, url) | Dictionary
reactions | reakcije na sporoilo | Dictionary
mentions | kdo vse je omenjen v sporoilu | Dictionary
reference | referenca na drugo sporoilo | Dictionary 

### UREJANJE IZVORNIH PODATKOV
----- 

Za pridobitev podatkov z Discord strenika smo uporabili orodje `Discord chat exporter` (https://github.com/Tyrrrz/DiscordChatExporter/releases/tag/2.27). Podatkovna mnoica je sestavljena iz .JSON datotek. Te smo prebrali s knjinico `json`. Te datoteke smo nato z uporabo knjinice `pandas` pretvorili v DataFrame in iz tega naprej izvajali analozo podatkov. 


### ANALIZA PODATKOV IN UGOTOVITVE 
----- 

Pri pregledovanju sporoil, smo ugotovili, da tevilo sporoil je v povpreju mono prevladovalo v prvem letu uporabe serverja (tudijsko leto 2017/18), e naslednje leto pa je padlo skoraj na polovico. V tudijskem letu 2019/20 se je tevilo sporoil povealo pri predmetih iz drugega semestra, kar lahko pripisujemo pojavu koronavirusa. tudentje so morali poiskati drug vir komunikacije in oitno je bil Discord eden izmed njih.

![alt text](./Grafi/etrta.png?raw=true)


tevilo pripetih sporoil, ki ponavadi predstavljajo snov, se je z leti precej zmanjalo. To ponazarja, da veliko tudentov e vedno uporablja snov preteklih let. Pri nekaterih predmetih se je z leti prenehalo dodajanje snovi. Iz tega lahko razberemo, da se tudi snov pri posameznih predmetih ne spreminja veliko.



Pri analizi reakcij na sporoila smo ugotovil, da se v drugem semestru reakcije oz. emjoi-ji pogosteje in vekrat uporabljajo. e primerjamo uporabo najvekrat uporabljenega emojia v prvem semestru, je ta 157 uporab, v drugem semestru pa kar 500 uporab. V obeh semestrih, se najpogosteje uporablja emoji :lekosalute:.

Po naem mnenju, se v drugem semestru bolj uporabljajo emoji-ji, zato ker ima Discord takrat ve uporabnikov kot v prvem semestru. Na zaetku prvega semestra marsi kdo sploh ne ve za ta Discord server, ki ga imamo, zato ga tudi ne uporablja, medtem ko v drugem semestru, Discord server veina tudentov e pozna in ga tudi uporablja.



 


Iskali smo frekvenco kletvic v prvem letniku za vse predmete. Za pomo smo uporabljali knjinici `profanity_filter` ter `spacy`, ki z NLP iejo kletvice v stavkih. Iz grafov opazimo, da se tevilo pri prvih petih predmetih mono povea ob koncu januarja, pri drugih petih predmetih pa ob koncu junija torej v asu izpitnega obdobja.

Iskali smo uporabnika s najve pripetimi sporoili, saj s tem merimo kakovost sporoil. Uproabnik z najve pripetimi sporoili je `Kona`. To, da je sporoilo pripeto se smatra, kot kvalitetno sporoilo. Da je sporoilo pripeto, pomeni, da bo shranjeno v neko blinjico in bomo do njega lahko hitro dostopali (npr. neki zapiski...). Glede na to, da je `Kona` uporabnik z najve pripetimi sporoilo, lahko smatramo, da je to uporabnik, ki napie najbolj kvalitetna sporoila, ki bodo tudentom kasneje prila prav.
