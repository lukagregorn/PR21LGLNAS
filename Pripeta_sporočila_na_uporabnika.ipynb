## Pripeta sporočila na uporabnika

Prikazal vam bom kateri uporabniki imajo največ pripetih sporočil (s tem lahko merimo kvaliteto sporočil, saj ta sporočila ponavadi vsebujejo pomembno vsebino za predmet). 

**Priprava podatkov**

```python
# uporabljene knjižnice
import json
import pandas as pd
```

```python
# prvi semester
matematika = json.load(open('DATA/matematika.json', encoding="utf8"))
matematikaDf = pd.DataFrame(matematika["messages"]) 

diskretne = json.load(open('DATA/diskretne-strukture.json', encoding="utf8"))
diskretneDf = pd.DataFrame(diskretne["messages"]) 

programiranje1 = json.load(open('DATA/programiranje-1.json', encoding="utf8"))
programiranje1Df = pd.DataFrame(programiranje1["messages"]) 

uvod = json.load(open('DATA/uvod-v-racunalnistvo.json', encoding="utf8"))
uvodDf = pd.DataFrame(uvod["messages"]) 

arhitektura = json.load(open('DATA/racunalniska-arhitektura.json', encoding="utf8"))
arhitekturaDf = pd.DataFrame(arhitektura["messages"]) 
```

```python
# drugi semester 
komunikacije = json.load(open('DATA/racunalniske-komunikacije.json', encoding="utf8"))
komunikacijeDf = pd.DataFrame(komunikacije["messages"]) 

operacijski = json.load(open('DATA/operacijski-sistemi.json', encoding="utf8"))
operacijskiDf = pd.DataFrame(operacijski["messages"]) 

programiranje2 = json.load(open('DATA/programiranje-2.json', encoding="utf8"))
programiranje2Df = pd.DataFrame(programiranje2["messages"]) 

podatkovneBaze = json.load(open('DATA/podatkovne-baze.json', encoding="utf8"))
podatkovneBazeDf = pd.DataFrame(podatkovneBaze["messages"]) 

statistika = json.load(open('DATA/statistika.json', encoding="utf8"))
statistikaDf = pd.DataFrame(statistika["messages"]) 
```

**Koda:**

```python
# najprej ustvarimo slovar, v katerega bomo vstavljali kot ključ uporabnika, 
# kot vrednost mu bomo pa prištevali število pripetih sporočil 

uporabniki = {} 

# za vsak predmet v prvem semestru poiščemo uporabnike s najbolj pripetimi sporočili 

for index, row in matematikaDf.iterrows(): 
    if (row["isPinned"] == True): 
        if (row["author"]["name"] not in uporabniki): 
            uporabniki.update({ row["author"]["name"] : 1}) 
        else: 
            uporabniki[row["author"]["name"]] += 1
```

```python
import pandas as pd

# izpis uporabnikov in njihovih pripetih sporočil
df1 = pd.DataFrame(list(uporabniki.items())) 
df1.columns = ["author", "num_of_pins"]
df1
```

Uporabnike, ki imajo `2` pripeta sporočila ali manj bom odstranil, saj iščemo uporabnika s največ pripetimi sporočili ter uporabnikov, ki imajo samo `1` ali `2` pripeta sporočila je zelo veliko. 

```python
uporabniki1 = {} 

for key, value in uporabniki.items(): 
    if (value > 2): 
        uporabniki1.update({key : value})
```

**Izpis uporabnikov, ki imajo `3` ali več pripetih sporočil:**

```python
# izpis uporabnikov in njihovih pripetih sporočil
df2 = pd.DataFrame(list(uporabniki1.items())) 
df2.columns = ["author", "num_of_pins"]
df2
```

**Graf uporabnikov in njihovih pripetih sporočil**

```python
%matplotlib inline
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize=(28,25)) 

plt.barh(*zip(*uporabniki1.items()), height=0.7, color="purple") 

plt.title("Pripeta sporočila glede na uporabnike", fontsize=40) 
plt.xlabel("Število pripetih sporočil", fontsize=30)
plt.ylabel("Uporabniki", fontsize=30) 
plt.xticks(fontsize=30) 
plt.yticks(fontsize=30)
plt.show()
```

**Uporabnik z največ pripetimi sporočili:**

```python
import operator

# poiščemo uporabnika z največ pripetimi sporočili
user = max(uporabniki1.items(), key=operator.itemgetter(1))[0]
user
```

Uproabnik z največ pripetimi sporočili je `Kona`. To, da je sporočilo pripeto se smatra, kot kvalitetno sporočilo. Da je sporočilo pripeto, pomeni, da bo shranjeno v neko bližnjico in bomo do njega lahko hitro dostopali (npr. neki zapiski...). Glede na to, da je `Kona` uporabnik z največ pripetimi sporočilo, lahko smatramo, da je to uporabnik, ki napiše najbolj kvalitetna sporočila, ki bodo študentom kasneje prišla prav. 

Pri raziskavi, smo naleteli, da ima ogromno študentov le eno pripeto sporočilo, še več je pa takih, ki pa nimajo nobenega sporočila pripetega.
