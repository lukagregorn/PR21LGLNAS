## Najpogosteje reakcije na sporoƒçila

Prikazal vam bom najpogosteje reakcije na sporoƒçila.

**Priprava podatkov**

```python
# uporabljene knji≈ænice
import json
import pandas as pd
```

```python
# prvi semester
matematika = json.load(open('DATA/matematika.json', encoding="utf8"))
matematikaDf = pd.DataFrame(matematika["messages"]) 

diskretne = json.load(open('DATA/diskretne-strukture.json', encoding="utf8"))
diskretneDf = pd.DataFrame(diskretne["messages"]) 

programiranje1 = json.load(open('DATA/programiranje-1.json', encoding="utf8"))
programiranje1Df = pd.DataFrame(programiranje1["messages"]) 

uvod = json.load(open('DATA/uvod-v-racunalnistvo.json', encoding="utf8"))
uvodDf = pd.DataFrame(uvod["messages"]) 

arhitektura = json.load(open('DATA/racunalniska-arhitektura.json', encoding="utf8"))
arhitekturaDf = pd.DataFrame(arhitektura["messages"]) 
```

```python
# drugi semester 
komunikacije = json.load(open('DATA/racunalniske-komunikacije.json', encoding="utf8"))
komunikacijeDf = pd.DataFrame(komunikacije["messages"]) 

operacijski = json.load(open('DATA/operacijski-sistemi.json', encoding="utf8"))
operacijskiDf = pd.DataFrame(operacijski["messages"]) 

programiranje2 = json.load(open('DATA/programiranje-2.json', encoding="utf8"))
programiranje2Df = pd.DataFrame(programiranje2["messages"]) 

podatkovneBaze = json.load(open('DATA/podatkovne-baze.json', encoding="utf8"))
podatkovneBazeDf = pd.DataFrame(podatkovneBaze["messages"]) 

statistika = json.load(open('DATA/statistika.json', encoding="utf8"))
statistikaDf = pd.DataFrame(statistika["messages"]) 
```

**Priprava reakcij (1. semester)**

```python
reakcije = {} 

for index, row in matematikaDf.iterrows(): 
    if len(row["reactions"]) != 0:
        for vrstica in row["reactions"]:
            if vrstica['emoji']['name'] not in reakcije:
                reakcije.update({vrstica['emoji']['name'] : 1}) 
            else: 
                reakcije[vrstica['emoji']['name']] += 1
```

```python
for index, row in diskretneDf.iterrows(): 
    if len(row["reactions"]) != 0:
        for vrstica in row["reactions"]:
            if vrstica['emoji']['name'] not in reakcije:
                reakcije.update({vrstica['emoji']['name'] : 1}) 
            else: 
                reakcije[vrstica['emoji']['name']] += 1
```

```python
for index, row in programiranje1Df.iterrows(): 
    if len(row["reactions"]) != 0:
        for vrstica in row["reactions"]:
            if vrstica['emoji']['name'] not in reakcije:
                reakcije.update({vrstica['emoji']['name'] : 1}) 
            else: 
                reakcije[vrstica['emoji']['name']] += 1
```

```python
for index, row in uvodDf.iterrows(): 
    if len(row["reactions"]) != 0:
        for vrstica in row["reactions"]:
            if vrstica['emoji']['name'] not in reakcije:
                reakcije.update({vrstica['emoji']['name'] : 1}) 
            else: 
                reakcije[vrstica['emoji']['name']] += 1
```

```python
for index, row in arhitekturaDf.iterrows(): 
    if len(row["reactions"]) != 0:
        for vrstica in row["reactions"]:
            if vrstica['emoji']['name'] not in reakcije:
                reakcije.update({vrstica['emoji']['name'] : 1}) 
            else: 
                reakcije[vrstica['emoji']['name']] += 1
```

```python
# reakcije (emoji-i) uporabljeni v prvem semestru 
reakcije
```

Ker gledamo najpogosteje uporabljene reakcije, sem tiste, ki so uporabljene manj kot 20-krat odstranil, saj ima najpogosteje uporabljena reakcija v tem semestru 157 uporab.

```python
zbReakcije = {}
for key, value in reakcije.items(): 
    if (value >= 20): 
        if key == 'üòÇ':
            zbReakcije.update({':joy:' : value})
        elif key == 'üëå': 
            zbReakcije.update({':ok:' : value})
        elif key == 'üçÜ': 
            zbReakcije.update({':eggplant:' : value}) 
        elif key == '‚ù§':
            zbReakcije.update({':heart:' : value}) 
        elif key == 'üëç':
            zbReakcije.update({':thumbsup:' : value}) 
        elif key == 'üíØ':
            zbReakcije.update({':hundredpoints:' : value}) 
        elif key == '‚úÖ': 
            zbReakcije.update({':whiteheavycheckmark:' : value}) 
        elif key == 'üóë': 
            zbReakcije.update({':wastebasket:' : value}) 
        elif key == '‚ùå': 
            zbReakcije.update({':crossmarkemoji:' : value}) 
        elif key == '‚ù§Ô∏è': 
            zbReakcije.update({':redheartemoji:' : value}) 
        else: 
            zbReakcije.update({key : value}) 
            
```

```python
# izbolj≈°an slovar reakcij
zbReakcije
```

**Grafiƒçni prikaz podatkov**

```python
%matplotlib inline
import matplotlib.pyplot as plt

csfont = {'fontname':'DejaVu Sans'}

f, ax = plt.subplots(figsize=(27,25)) 

plt.barh(*zip(*zbReakcije.items()), height=0.7, color="darkgreen") 
vr = 0
for key, value in zbReakcije.items():
    plt.text(value, vr, str(value), fontsize=20)
    vr+=1
plt.title("Porazdelitev reakcij na sporoƒçila (1. semester)", fontsize=40) 
plt.xlabel("≈†tevilo reakcij", fontsize=30)
plt.ylabel("Reakcija", fontsize=30) 
plt.xticks(fontsize=20, **csfont) 
plt.yticks(fontsize=20)
plt.show()
```
