```python
# prvi semester
matematika = json.load(open('DATA/matematika.json', encoding="utf8"))
matematikaDf = pd.DataFrame(matematika["messages"]) 

diskretne = json.load(open('DATA/diskretne-strukture.json', encoding="utf8"))
diskretneDf = pd.DataFrame(diskretne["messages"]) 

programiranje1 = json.load(open('DATA/programiranje-1.json', encoding="utf8"))
programiranje1Df = pd.DataFrame(programiranje1["messages"]) 

uvod = json.load(open('DATA/uvod-v-racunalnistvo.json', encoding="utf8"))
uvodDf = pd.DataFrame(uvod["messages"]) 

arhitektura = json.load(open('DATA/racunalniska-arhitektura.json', encoding="utf8"))
arhitekturaDf = pd.DataFrame(arhitektura["messages"]) 
```

**Priprava reakcij**

```python
reakcije = {} 

for index, row in matematikaDf.iterrows(): 
    if len(row["reactions"]) != 0:
        for vrstica in row["reactions"]:
            if vrstica['emoji']['name'] not in reakcije:
                reakcije.update({vrstica['emoji']['name'] : 1}) 
            else: 
                reakcije[vrstica['emoji']['name']] += 1
```

```python
for index, row in diskretneDf.iterrows(): 
    if len(row["reactions"]) != 0:
        for vrstica in row["reactions"]:
            if vrstica['emoji']['name'] not in reakcije:
                reakcije.update({vrstica['emoji']['name'] : 1}) 
            else: 
                reakcije[vrstica['emoji']['name']] += 1
```

```python
for index, row in programiranje1Df.iterrows(): 
    if len(row["reactions"]) != 0:
        for vrstica in row["reactions"]:
            if vrstica['emoji']['name'] not in reakcije:
                reakcije.update({vrstica['emoji']['name'] : 1}) 
            else: 
                reakcije[vrstica['emoji']['name']] += 1
```

```python
for index, row in uvodDf.iterrows(): 
    if len(row["reactions"]) != 0:
        for vrstica in row["reactions"]:
            if vrstica['emoji']['name'] not in reakcije:
                reakcije.update({vrstica['emoji']['name'] : 1}) 
            else: 
                reakcije[vrstica['emoji']['name']] += 1
```

```python
for index, row in arhitekturaDf.iterrows(): 
    if len(row["reactions"]) != 0:
        for vrstica in row["reactions"]:
            if vrstica['emoji']['name'] not in reakcije:
                reakcije.update({vrstica['emoji']['name'] : 1}) 
            else: 
                reakcije[vrstica['emoji']['name']] += 1
```

```python
# reakcije (emoji-i) uporabljeni v prvem semestru 
reakcije
```

Ker gledamo najpogosteje uporabljene reakcije, sem tiste, ki so uporabljene manj kot 20-krat odstranil.

```python
zbReakcije = {}
for key, value in reakcije.items(): 
    if (value >= 20): 
        zbReakcije.update({key : value})
```

```python
# zbolj≈°an slovar reakcij
zbReakcije
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

csfont = {'fontname':'DejaVu Sans'}

f, ax = plt.subplots(figsize=(27,25)) 

plt.barh(*zip(*zbReakcije.items()), height=0.7, color="darkgreen") 
vr = 0
for key, value in zbReakcije.items():
    plt.text(value, vr, str(value), fontsize=20)
    vr+=1
plt.title("Porazdelitev reakcij na sporoƒçila (1. semester)", fontsize=30) 
plt.xlabel("≈†tevilo reakcij", fontsize=25)
plt.ylabel("Reakcija", fontsize=25) 
plt.xticks(fontsize=20, **csfont) 
plt.yticks(fontsize=20)
plt.show()
```
